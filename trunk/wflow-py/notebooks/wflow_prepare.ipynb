# Introduction

This notebook shows the steps to prepate the maps needed fro the wflow_* set op models. The wflow_prepare scripts have a numebr of specific feutures that are not used here. This notebook shows the principles and may be used as a template to create better preprocessing options.

In short the steps to generate input files for the wflow models are:

- Create a DEM for the catchment
- Create a local crainage network file for the catchment used the DEM and additional information
- Create a subarea/catchment map by indicating outflow points (gauges) in the map
- Convert/resample landuse and soil maps


In this example we have the following maps at our disposal:

- a 90x90m (approx) resolution DEM based on SRTM
- a ??x?? resolution soil map (FAO global map)
- a ??x?? resolution landuse map (corine)

## Example

### First try using the original DEM

Begin by importing the wflow_lib module. This module also imports the pcraster framework.

```python
import wflow.wflow_lib as tr
```

start by reading the DEM into memory. Note that a DEM can be rather large, in that case it may be good to resample it forst before reading it into memory.

```python
dem = tr.readmap("example_dem.map")
```

Check the dem by plotting it. Note the use of the _pcr2numpy_ function to convert the pcraster map object to a numpy matrix

```python
imshow(tr.pcr2numpy(dem,np.nan))
colorbar()
```

```python
ldd = tr.lddcreate_save("ldd.map",dem,True)
```

We have now created a first version of the LDD. We can plot it with python but unfortunately it does not show the flow directiosn as nicely as the specialised pcraster graphics (aguila). We can however, use one of the pcraster functions (streamorder) to show how the main rivers run:

```python
strorder = tr.streamorder(ldd)
imshow(tr.pcr2numpy(strorder,np.nan))
```

wflow_lib has a function to search for the largest catchment within the current DEM. It returns a map with a single cell filled at the point of the outlet. Next we use that map to determine the whoel catchment area:

```python
outletpoint = tr.find_outlet(ldd)
catch = tr.subcatch(ldd,outletpoint)
imshow(tr.pcr2numpy(catch,np.nan))
```

In this case we get something that is not  how we want it to be. In the norther and easter part of the map the catchment delineation seesm to be missing stuff as we can be from the map below (which is more-or-less correct). The reason in this case is that a 4x4km DEM is missing some of the feature to correctly deliniate the DEM. One way to correct this is to use a know river course shapefile and use that to "burn in" the river into the DEm and re-create the LDD.

```python
goodcatch = tr.readmap("wflow_subcatch.map")
imshow(tr.pcr2numpy(goodcatch,np.nan))
```

### Second try, use a river network

In this case we use a grid (converted from a shape-file, best to use qgis to do that) to burn-in the river network (only the higher orders)

```python
river = tr.readmap("river.map")
imshow(tr.pcr2numpy(river,np.nan))
```

```python
tr.setglobaloption("unitcell")
demregional=tr.windowaverage(dem,100)
ldddem = tr.ifthenelse(tr.scalar(river) >= 1.0, demregional -9000 , dem)
```

```python
ldd = tr.lddcreate_save("ldd.map",ldddem,True)
strorder = tr.streamorder(ldd)
imshow(tr.pcr2numpy(strorder,np.nan))
```

```python
outletpoint = tr.find_outlet(ldd)
catch = tr.subcatch(ldd,outletpoint)
imshow(tr.pcr2numpy(catch,np.nan))
```
